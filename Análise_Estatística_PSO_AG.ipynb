# Análise Estatística da Metaheurística de Enxame de Partículas

A metaheurística de enxame de partículas foi desenvolvida em Matlab e aplicada ao problema de despacho econômico com efeito de ponto de carregamento de válvula. A função objetivo representa o custo de produção de energia. A  restrição de igualdade representa o balanço de potência ativa, e visa garantir que a potência gerada total seja igual à potência demandada. As restrições de desigualdade são restrições operacionais cujo objetivo é garantir o funcionamento das unidades geradoras dentro de seus limites mínimo e máximo de geração. A formulação apresentada não representa as restrições operacionais relacionadas ao sistema de transmissão e também despreza as perdas neste sistema. Assim, sendo, neste sentido, o PDE deste trabalho é uma formulação simplificada para o problema de despacho. Por apresentar o efeito de ponto de carregamento de válvula, a função objetivo do problema torna-se não diferenciável em tais pontos. Além disso, também é não-convexa e possui vários pontos de mínimos locais. Todas as variáveis do problema são contínuas.

```python
import pandas as pd
import numpy as np
import scipy as spy
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

Foram realizadas 100 execuções de 100 iterações cada, do algoritmo de enxame de partículas. Em todos os casos utilizou-se o número de 40 partículas. Os resultados são exibidos a seguir:

## - Geração com 40 Geradores

```python
best = pd.read_csv('MPGB40.csv',header=None).iloc[:,:].values

custo = pd.read_csv('MC40.csv',header=None).iloc[:,:].values
```

```python
plt.figure(figsize=(16,8))

for execucao in range(100):
    
    sns.lineplot(x = np.arange(1,100), y=custo[execucao,1:])


plt.title('Custo de Geração (40 Geradores)')
plt.xlabel('Iteração')
plt.ylabel('Custo ($)')

    
```

```python
print('Média do Custo Otimizado: ', np.round(np.mean(custo[:,-1]),4),'$\n')
print('Desvio Padrão do Custo Otimizado: ', np.round(np.std(custo[:,-1]),4),'$\n')
print('Valor Mínimo Obtido Custo Otimizado: ', np.round(np.min(custo[:,-1]),40),'$\n')
print('Valor Máximo Obtido Custo Otimizado: ', np.round(np.max(custo[:,-1]),40),'$\n')
```

```python
plt.figure(figsize=(16,4))

sns.boxplot(custo[:,-1])
```

```python
print('Melhor Solução: ', best[np.argmin(custo[:,-1])],'MW\n')

print('Demanda Atendida: ', np.sum(best[np.argmin(custo[:,-1])]),'MW\n')
```
